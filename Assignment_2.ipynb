1. Write a Python program that takes an integer as input and calculates the
sum of its digits


```python
num = int(input("Enter any two digit number: "))
total_sum = 0

temp = abs(num) # helps to change negative number to positive
while temp > 0:
  digit = temp % 10
  total_sum += digit
  temp //= 10
print(f"sum of digit: {total_sum}")
```

2. Create a Python program to calculate the factorial of a given number using
a function

3. Develop a Python program that checks whether a given word or phrase is a
palindrome.

```python
def palindrome(text):
  text = text.replace(" ","").lower()
  return text == text[::-1]

word= input("Enter the word: ")

if palindrome(word):
  print("It is palindrome")
else:
  print("It is not palindrome")
```

4. Create a Python program that takes a sentence as input and counts the
number of vowels and consonants.

```python
sentence = input("Enter a word or sentence : ")
vowels ="aeiou"
t_vowels = 0
t_consonats = 0
for char in sentence.lower():
  if char.isalpha(): # aplha ignore space and takes only letters
    if char in vowels:
      t_vowels += 1
    else:
      t_consonats += 1
print(f"Total numbers of vowels and consonats: {t_vowels} , {t_consonats}")
```

7. Develop a Python program that checks if a given year is a leap year.

```python
year = int(input("Enter a year: "))
if (year % 4 == 0):
    if (year % 100 == 0):
        if (year % 400 == 0):
            print(year, "is a leap year.")
        else:
            print(year, "is not a leap year.")
    else:
        print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

9. Write a Python program to generate the Fibonacci sequence up to a
specified number of terms.

```python
def fib(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b

num = int(input("Enter the number of terms: "))
fib(num)
```

10. Write a python program to find the roots of a quadratic equation


```python
a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))

D = b**2 - 4*a*c

if D > 0:
    root1 = (-b + (D ** 0.5)) / (2 * a)
    root2 = (-b - (D ** 0.5)) / (2 * a)
    print("Two real roots:", root1, "and", root2)

elif D == 0:
    root = -b / (2 * a)
    print("One real root:", root)

else:
    real = -b / (2 * a)
    imaginary = ((-D) ** 0.5) / (2 * a)
    print("Complex roots:")
    print(f"{real} + {imaginary}i")
    print(f"{real} - {imaginary}i")
```

4. Write a Python function that takes a positive integer and returns the sum
of the cube of all the positive integers smaller than the specified number.
Sample input: 4 Sample output: 36.

```python
def p_int(n):
  result = 0
  for i in range(1,n):
    num = i**3
    result += num
  return result

p_num = int(input("Enter the number "))
print(f"Output: {p_int(p_num)}")
```

6. Write a Python program to find the most frequent item in a list of
numbers. Sample input: 2, 3, 4, 2, 5, 2 Sample output: 2

```python
freq_num = input("Enter the numbers separated by commas: ")
num_list = list(map(int, freq_num.split(",")))# Convert input into list of integers

frequency = {}

for n in num_list:
    if n in frequency:
        frequency[n] += 1
    else:
        frequency[n] = 1
most_frequent = max(frequency, key=frequency.get)

print("Most frequent number is:", most_frequent)
```

7. Write a Python program to find the sum of squares of the numbers in a
list. Sample input: 2,1,3,1 Sample output: 15

```python
def s_square(n):
  result = 0
  for i in n:
    num = i**2
    result += num
  return result
num = input("Enter the numbers using commas: ")
num_list = list(map(int, num.split(",")))
print(f"output: {s_square(num_list)}")
```
