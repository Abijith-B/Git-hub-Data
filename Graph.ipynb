```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris
```

#For Visualizing Single Variables (Univariate Analysis)

```python
sns.histplot(iris['petal_length'])
```

```python
sns.kdeplot(iris['petal_length'],fill = True)
```

```python
sns.countplot(x='species', data = iris)
```

#For Visualizing Two Variables (Bivariate Analysis)

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
iris.plot.scatter(x='sepal_length', y='sepal_width')
```

```python
sns.jointplot(x= 'sepal_length', y = 'sepal_width', data = iris, kind = 'scatter')
```

```python
sns.lineplot(x='sepal_length',y='sepal_width',data= iris)
```

#For Comparing Distributions Across Categories

```python
sns.boxplot(x='species',y='petal_length',data = iris)
```

```python
sns.violinplot(x='species',y='petal_length',data=iris)
```

```python
sns.swarmplot(x='species',y='petal_length',data = iris,size = 3)
```

#For Visualizing Multivariate Data

```python
corr = iris.corr(numeric_only = True)
sns.heatmap(corr, annot = True, cmap = 'coolwarm')
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(iris,figsize = (6,6),diagonal= 'kde')
```
