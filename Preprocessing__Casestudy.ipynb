#Duplicate Removal

#Handling Missing Values

```python
column_missing_value=house_price.isnull().sum()
print(column_missing_value)
```

```python
numerical_values=house_price.select_dtypes(include=['int64','float64'])
house_price[numerical_values.columns]=numerical_values.fillna(numerical_values.median())
house_price[numerical_values.columns].isnull().sum()
```

```python
house_price['No of Times Visited'] = house_price['No of Times Visited'].fillna(house_price['No of Times Visited'].mode()[0])
house_price['No of Times Visited'].isnull().sum()
```

```python
cate_values=house_price.select_dtypes(include=['object'])
house_price[cate_values.columns]=cate_values.fillna(cate_values.mode())
house_price[cate_values.columns].isnull().sum()
```

#Outlier Removal

```python
num_cols = house_price.select_dtypes(include=["int64", "float64"]).columns

outlier_summary = {}

for col in num_cols:
    Q1 = house_price[col].quantile(0.25)
    Q3 = house_price[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = house_price[(house_price [col] < lower_bound) | (house_price [col] > upper_bound)]
    outlier_summary[col] = outliers.shape[0]

outlier_summary
```

```python
features_to_clean = [
    "Sale Price",
    "Flat Area (in Sqft)",
    "Lot Area (in Sqft)",
    "Living Area after Renovation (in Sqft)",
    "Lot Area after Renovation (in Sqft)"
]

# Apply IQR-based outlier removal
House_Pricing_cleaned = house_price.copy()

for col in features_to_clean:
    Q1 = House_Pricing_cleaned[col].quantile(0.25)
    Q3 = House_Pricing_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    House_Pricing_cleaned = House_Pricing_cleaned[
        (House_Pricing_cleaned [col] >= lower_bound) &
        (House_Pricing_cleaned [col] <= upper_bound)
    ]

# Dataset size comparison
print("Original shape:", house_price.shape)
print("After outlier removal:", House_Pricing_cleaned.shape)
```

#Scaling Numerical Variables

```python
from sklearn.preprocessing import StandardScaler

num_features = [
    col for col in House_Pricing_cleaned.columns
    if col != "Sale Price" and House_Pricing_cleaned[col].dtype in ["int64", "float64"]
]

print("Numerical features to scale:")
print(num_features)

scaler = StandardScaler()

House_Pricing_cleaned[num_features] = scaler.fit_transform(
    House_Pricing_cleaned[num_features]
)

print("Scaling completed successfully")
```

```python
num_feature = [
    col for col in House_Pricing_cleaned.columns
    if col != "Sale Price" and House_Pricing_cleaned[col].dtype in ["int64", "float64"]
]

# Apply Standard Scaling
standard_scaler = StandardScaler()
House_Pricing_cleaned[num_features] = standard_scaler.fit_transform(
    House_Pricing_cleaned[num_features]
)
print("Standard scaling applied successfully")

House_Pricing_cleaned[num_feature].head()
```
