#QN 1
● Use https://api.spacexdata.com/v4/launches


● Extract relevant columns: name, date_utc, success, details, rocket


● Convert date_utc to datetime and extract the year


```python
import requests
import pandas as pd
import numpy as np
import sqlite3
```

```python
url = "https://api.spacexdata.com/v4/launches"# Load SpaceX launch data from API
data = requests.get(url)
data1 = data.json()

df = pd.DataFrame(data1, columns=["name", "date_utc", "success", "details", "rocket"])# Create a DataFrame with required columns
print(df)
```

```python
df["date_utc"] = pd.to_datetime(df["date_utc"])# Convert date_utc to datetime format
df["year"] = df["date_utc"].dt.year# Extract year from the date
print(df)
```


#QN 2
● Use https://api.spacexdata.com/v4/rockets


● Extract id, name, type, active, and stages


```python
url = "https://api.spacexdata.com/v4/rockets"
rocket_data = requests.get(url).json()
rocket_df = pd.DataFrame(rocket_data, columns=["id", "name", "type", "active", "stages"])

print(rocket_df)
```

#QN 3

● Join the two DataFrames on rocket ID using

```python
merge_df = pd.DataFrame(df.merge(rocket_df, left_on="rocket", right_on="id"))
merge_df
```

#QN4

● Add a new column country and randomly assign one of these values:
['USA', 'Russia', 'India', 'China', 'France']

```python
countries = ['USA', 'Russia', 'India', 'China', 'France']
merge_df["country"] = np.random.choice(countries, size=len(merge_df))

merge_df
```

#QN 5

● Use sqlite3 to create a connection and save the merged DataFrame as a table named launches


● Table should contain all merged columns including country

```python
conn = sqlite3.connect("spacex.db")
merge_df.to_sql("launches", conn, if_exists="replace", index=False)
conn.close()
merge_df
```

#QN 6
1. Launches by Country

2. Which year had the highest number of launches?

3. Top 5 Missions by Launch Count

4. Top Missions by rocket Count

```python
conn = sqlite3.connect('spacex.db')

query = """
SELECT country, COUNT(*) AS launch_count
FROM launches
GROUP BY country
ORDER BY launch_count DESC;
"""
df_country = pd.read_sql_query(query, conn)
df_country
```

```python
query = """
SELECT year, COUNT(*) AS launch_count
FROM launches
GROUP BY year
ORDER BY launch_count DESC
LIMIT 1;
"""
df_top_year = pd.read_sql_query(query, conn)
df_top_year
```

```python
query = """
SELECT name_x, COUNT(*) AS launch_count
FROM launches
GROUP BY name_x
ORDER BY launch_count DESC
LIMIT 5;
"""

df_top_missions = pd.read_sql_query(query, conn)
df_top_missions
```

```python
query = """
SELECT name_y, COUNT(*) AS launch_count
FROM launches
GROUP BY name_y
ORDER BY launch_count DESC
LIMIT 1;
"""

df_top_missions = pd.read_sql_query(query, conn)
df_top_missions
```
