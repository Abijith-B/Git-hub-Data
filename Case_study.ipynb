# Q1. Write a function that return the range (maximum and minimun) of a given list of numbers

```python
def range(list_number):
  minimum=list_number[0]#setting the first value of the list as minimum
  maximum=list_number[0]#setting the first value of the list as maximum

  for num in list_number:#looping through the entire list for finding the maximum and minimum value
    if (num <= minimum):
      minimum=num

    if(num >= maximum):
      maximum=num

  return(minimum,maximum)

my_list=[10,25,21,41,8,36]
value_range=range(my_list)#value of the range funtion is stored
value_range
```

# Q2. Write function that returns mean value of a list of numbers

```python
def mean(n):
  total=0
  for i in n:
    total=total+i#or total+=i

  return total/len(n)

my_list=[5,10,15,20,25,30,35]
mean_value=mean(my_list)
mean_value
```

# Q3. Write a function that returns the median value of a list of number

```python
def median(number):
  number=sorted(number)#sorting the numbers
  n=len(number)#taking the length
  middle=n//2#to find the middle value

  if n%2==0:#function top check even or odd
    return (number[middle-1]+number[middle])/2#[] used to denote the index of the given value (here middle)
  else:
    return number[middle]#if odd return the value

my_list=[2,1,3,5,4,6,7,8]
num=median(my_list)
num
```
